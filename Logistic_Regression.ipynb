```python
# Importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix

data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Datasets/framingham.csv')
data.head()
```

```python
# Partitioning data for training
Predictors = data.drop(columns=['TenYearCHD']).columns.tolist()
x = data[Predictors]
y = data['TenYearCHD']

# Training
x_train, x_val, y_train, y_val = train_test_split(x, y, test_size=0.2, random_state=1)
```

```python
# Fit a model
model = LogisticRegression(max_iter=2000)
model.fit(x_train, y_train)
```

```python
model.predict(x_val)
```

```python
# Validating the model
model.score(x_val, y_val)
```

```python
cm = confusion_matrix(y_val, model.predict(x_val))
cm
```

```python
plt.figure(figsize=(6, 4))
sns.heatmap(cm, annot=True, cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

```python
parameters = {
    'C': [0.001, 0.01, 0.1, 1, 10, 100, 1000],
    'max_iter': [100, 1000, 2500, 5000],
    'solver': ['lbfgs', 'liblinear', 'sag', 'saga']
}
lr = LogisticRegression()
```

```python
from sklearn.model_selection import GridSearchCV
grid_search = GridSearchCV(lr, parameters, cv=5)
grid_search.fit(x_train, y_train)
```

```python
grid_search.best_params_
```

```python
grid_search.best_score_
```

```python
grid_search.best_estimator_.predict(x_val)
```

```python
y_pred = grid_search.best_estimator_.predict(x_val)
cm2 = confusion_matrix(y_val, y_pred)
cm2
```
